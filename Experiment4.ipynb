Given Expression is : 
$X = ABC(D+E+F) + AC(C+D+E)$


The circuit would look like:      

```python
import wavedrom
wavedrom.render("""
{ "assign":[
  ["X",
    ["|",
      ["&", "A","B","C",["|","D","E","F"]],
      ["&", "A","C", ["|","D","C","E"]]
    ]
  ]
]}""")
```

Defining this fucntion: X=ABC(D+E+F)+AC(C+D+E)

```python
def func(A,B,C,D,E,F):
    return A and B and C and (D or E or F) or (A and C and (C or D or E))
```

```python
A=[0, 1, 0, 0, 1, 1, 0, 1]
E=[1, 0, 1, 0, 0, 1, 1, 0]
B=[0,1,1,0,1,0,1,1]
C=[1,0,1,1,0,1,0,1]
D=[1,1,0,0,1,0,0,0]
F=[1,1,0,0,0,1,0,1]
X=[func(A[i],B[i],C[i],D[i],E[i],F[i]) for i in range(len(A))]
```

```python
def waveform(bits, col='blue',ti=''):
  data = np.repeat(bits, 2) # repeat each bit in the array
  t = 0.5 * np.arange(len(data)) # define the time axis using the length of the array
  f = plt.figure(figsize=(10,1)) # change figure size
  plt.step(t, data, where='post', color=col,label=ti); # the actual plotting
  plt.gca().axis('off'); # remove the axes
  for tbit, bit in enumerate(bits): plt.text(tbit+0.5, -0.3, str(bit)) # print out the bits as strings

waveform(A,'blue','A')
waveform(B,'blue','B')
waveform(C,'blue','C')
waveform(D,'blue','D')
waveform(E,'blue','E')
waveform(F,'blue','F')
waveform(X,'red','X')
```
